## 多標籤問題 - Kaggle的競賽

### [Yelp Restaurant Photo Classification](https://www.kaggle.com/c/yelp-restaurant-photo-classification/overview)


資料來源為Yelp社群用戶上傳之餐廳照片，用戶會在上傳照片的同時給他們一些標註

目標是希望能夠過照片就知道這是一間怎麼樣的餐廳，總共有如下幾類:

>1. good for lunch

>2. good for dinner

>3. takes reservations

>4. outdoor seating

>5. restaurant is expensive

>6. has alcohol

>7. has table service

>8. ambience is classy

>9. good for kids

一間餐廳可能屬於不只一類，因此這個案例為多標籤問題

此競賽的評估指標為 Mean F1-Score
