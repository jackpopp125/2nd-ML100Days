## 重要知識點複習：如何選擇正確的啟動函數


### 根據各個函數的優缺點來配置

- 如果使⽤ ReLU，要⼩⼼設置 learning rate，注意不要讓網路出現很多「dead」 神經元，如果不好解決，可以試試 Leaky ReLU、PReLU 或者Maxout

### 根據問題的性質

- ⽤於分類器時，Sigmoid 函數及其組合通常效果更好

- 由於梯度消失問題，有時要避免使⽤ sigmoid 和 tanh 函數。ReLU 函數是⼀個通⽤的啟動函數，⽬前在⼤多數情況下使⽤

- 如果神經網路中出現死神經元，那麼 PReLU 函數就是最好的選擇

- ReLU 函數建議只能在隱藏層中使⽤

## Sigmoid

Sigmoid 數學函數表示方式

### $$ sigmoid = lambda x: 1 / (1 + np.exp(-x)) $$

```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline
```

```python
#Sigmoid 數學函數表示方式
#sigmoid = lambda x: 1 / (1 + np.exp(-x))
def sigmoid(x):
    return (1 / (1 + np.exp(-x)))

#Sigmoid 微分
def dsigmoid(x):
    return (x * (1 - x))

# linespace generate an array from start and stop value
# with requested number of elements. Example 10 elements or 100 elements.
x = plt.linspace(-10,10,100)

# prepare the plot, associate the color r(ed) or b(lue) and the label 
plt.plot(x, sigmoid(x), 'b', label='linspace(-10,10,10)')

# Draw the grid line in background.
plt.grid()

# 顯現圖示的Title
plt.title('Sigmoid Function')

# 顯現 the Sigmoid formula
plt.text(4, 0.8, r'$\sigma(x)=\frac{1}{1+e^{-x}}$', fontsize=15)

#resize the X and Y axes
plt.gca().xaxis.set_major_locator(plt.MultipleLocator(1))
plt.gca().yaxis.set_major_locator(plt.MultipleLocator(0.1))
 
# create the graph
plt.show()
```

## 作業目標:

寫出 ReLU & dReLU 一階導數

並列印

## 作業重點

Rectified Linear Unit- Relu

### $$ f(x)=max(0,x) $$

```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

'''
作業:
    寫出 ReLU & dReLU 一階導數
    並列印
'''


# ReLU 數學函數表示方式
# ReLU = lambda x: f(x)=max(0,x)
def ReLU(x):
    return ( abs(x) * (x > 0) )

# 微分 (ReLU 一階導數)
def dReLU(x):
    return ( 1 * (x > 0) )

# linespace generate an array from start and stop value
# with requested number of elements. Example 10 elements or 100 elements.
x = plt.linspace(-10,10,100)

# prepare the plot, associate the color r(ed) or b(lue) and the label 
plt.plot(x, ReLU(x), 'r', label='linspace(-10,10,10)')
plt.plot(x, dReLU(x), 'b', label='linspace(-10,10,10)')

# Draw the grid line in background.
plt.grid()

# 顯現圖示的Title
plt.title('ReLU, dReLU Function')


# create the graph
plt.show()
```

## 1. sigmoid函數

![img1-1](https://www.zhihu.com/equation?tex=sigmoid%28x%29%3D%5Cfrac%7B1%7D%7B1%2Be%5E%7B-x%7D%7D+)

## 2. tanh 函數

![img2-1](https://www.zhihu.com/equation?tex=tanh%28x%29%3D%5Cfrac%7Bsinh%28x%29%7D%7Bcosh%28x%29%7D%3D%5Cfrac%7B%5Cfrac%7Be%5Ez-e%5E%7B-z%7D%7D%7B2%7D%7D%7B%5Cfrac%7Be%5Ez%2Be%5E%7B-z%7D%7D%7B2%7D%7D%3D%5Cfrac%7Be%5Ez-e%5E%7B-z%7D%7D%7Be%5Ez%2Be%5E%7B-z%7D%7D)

![img2-2](https://www.zhihu.com/equation?tex=%28tanh%28x%29%29%27%3D1-tanh%5E2%28x%29)

```python
x =  np.arange(-10,10,0.01)
y=(np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))

# tanh
plt.plot(x,y)
plt.grid(color='b' , linewidth='0.1' ,linestyle='--')

# tanh 一階導數
dev_y=1-np.square(y)
plt.plot(x,dev_y)
plt.grid(color='b' , linewidth='0.1' ,linestyle='--')
```

## 3. ReLU(Rectified Linear Unit 修正的線性單元)

![img3-1](https://www.zhihu.com/equation?tex=%5Cbegin%7Bequation%7D+f%28x%29%3D%5Cleft%5C%7B+%5Cbegin%7Baligned%7D+x+%26+%26+if+%5Cquad+x%5Cgeq0+%5C%5C0+%26+%26+if+%5Cquad+x%3C0+%5Cend%7Baligned%7D+%5Cright.+%5Cend%7Bequation%7D)

![img3-2](https://www.zhihu.com/equation?tex=%5Cbegin%7Bequation%7D+f%28x%29%E2%80%99%3D%5Cleft%5C%7B+%5Cbegin%7Baligned%7D+1+%26+%26+if+%5Cquad+x%5Cgeq0+%5C%5C0+%26+%26+if+%5Cquad+x%3C0+%5Cend%7Baligned%7D+%5Cright.+%5Cend%7Bequation%7D)

```python
import numpy as np
import matplotlib.pyplot as plt
 
x =  np.arange(-10,10,0.01)
y=list(map(lambda x:x if x>0 else 0,x))

# ReLU 
plt.plot(x,y)
plt.grid(color='b' , linewidth='0.1' ,linestyle='--')

# ReLU 一階導數
dev_y =list(map(lambda x: 1 if x>0 else 0,x))
plt.plot(x,dev_y)
plt.grid(color='b' , linewidth='0.1' ,linestyle='--')
```

## 4. leaky ReLU（帶洩露的ReLU）

![img4-1](https://www.zhihu.com/equation?tex=%5Cbegin%7Bequation%7D+f%28x%29%3D%5Cleft%5C%7B+%5Cbegin%7Baligned%7D+x+%26+%26+if+%5Cquad+x%5Cgeq0+%5C%5C+%5Clambda+x+%26+%26+if+%5Cquad+x%3C0+%5Cend%7Baligned%7D+%5Cright.+%5Cend%7Bequation%7D)

![img4-2](https://www.zhihu.com/equation?tex=%5Cbegin%7Bequation%7D+f%28x%29%27%3D%5Cleft%5C%7B+%5Cbegin%7Baligned%7D+1+%26+%26+if+%5Cquad+x%5Cgeq0+%5C%5C+%5Clambda+%26+%26+if+%5Cquad+x%3C0+%5Cend%7Baligned%7D+%5Cright.+%5Cend%7Bequation%7D)

```python
import numpy as np
import matplotlib.pyplot as plt
 
x =  np.arange(-10,10,0.01)
lambda1=0.1   # 指定 lambda 為 0.1
y=list(map(lambda x:x if x>0 else lambda1*x,x))

# leaky ReLU 
plt.plot(x,y)
plt.grid(color='b' , linewidth='0.1' ,linestyle='--')

# leaky ReLU 一階導數
dev_y = list(map(lambda x: 1 if x>0 else lambda1,x))
plt.plot(x,dev_y)
plt.grid(color='b' , linewidth='0.1' ,linestyle='--')
```
