## [度量聚类模型的好坏 - 轮廓系数](https://juejin.im/post/5b8f3bd06fb9a019d53ea30a)

这个指标度量模型将数据集分类的离散程度，即判断数据集是否分离的合理，

- 判断一个集群中的数据点是不是足够紧密（即内聚度），
- 一个集群中的点和其他集群中的点相隔是否足够远（即分离度），

故而轮廓系数结合了内聚度和分离度这两种因素，可以用来在相同原始数据的基础上用来评价不同算法，或者算法不同运行方式对聚类结果所产生的影响。

## x 是 高維空間中的一個點， w 是 高維空間中的一個向量，而 z 就是 x 對 w 的(內積)投影

我們希望經過「投影」之後，「讓投影後的 variance 越大越好」，希望投影後的 點的歧異度 仍然存在，所以慎選投影方向，讓variance 越大越好喔！
